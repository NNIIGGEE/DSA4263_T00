## SENTIMENT ANALYSIS USING VADER

```python
import pandas as pd
import nltk
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, classification_report
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from ast import literal_eval
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("preprocessed.csv", index_col=0, converters={'cleaned':literal_eval})
data
```

```python
data['cleaned2'] = data['cleaned'].apply(lambda x: ' '.join(x))
data
```

```python
le = LabelEncoder()
data["label"] = le.fit_transform(data["Sentiment"])
data
```

```python
model = SentimentIntensityAnalyzer()
data["predictedraw"] = data["cleaned2"].apply(lambda x: model.polarity_scores(x))
data
```

```python
def convert_score(raw):
    compound = raw["compound"]
    if compound >= 0: 
        score = 1
    else:
        score = 0
    return score
data["predicted"] = data["predictedraw"].apply(lambda row: convert_score(row))
data
```

```python
y_true = data["label"]
y_pred = data["predicted"]
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
auc = roc_auc_score(y_true, y_pred)

print("Accuracy = ",accuracy)
print("Precision = ", precision)
print("ROC AUC = ", auc)
print(classification_report(y_true, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

matrix = confusion_matrix(y_true, y_pred)

cm = ConfusionMatrixDisplay(confusion_matrix = matrix, display_labels = ["negative", "positive"])
cm.plot()
```
